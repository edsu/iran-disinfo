# Translate

```python
import pandas
import googletrans
```

In the Analysis notebook we discovered that AlinejadMasih was sent a lot of tweets by suspended accounts. However as you can below, most of these are in Arabic.

```python
pandas.set_option('display.max_colwidth', 280)

tweets = pandas.read_csv('results/AlinejadMasih.csv')
tweets.tweet_text
```

It's probably going to be far from perfect, but since I don't know Arabic I'm going to try to convert them to English using [googletrans](https://pypi.org/project/googletrans/) which hopefully will use the Google Translate to translate the Arabic into English.

```python
translator = googletrans.Translator()
translator.translate('تو خودت خدای تناقضی \nدنیا زن ایرانی را به نجابت و حیا می شناسد و اما تو برعکسی . آیا به تناقض تو نمی خندن', dest='en').text
```

Well, that might make sense? If it's not there alread there add a new column to the spreadsheet with the translated text. Note, we sleep because googletrans is scraping the Google Translate site, and we don't want to get blocked.

```python
import time
import random

translated = []
for text in tweets.tweet_text:
    try:
        translated.append(translator.translate(text, dest='en').text)
    except Exception as e:
        translated.append(None)
    time.sleep(random.randint(1, 5))

tweets['tweet_text_googletrans'] = pandas.Series(translated, index=tweets.index)               
```

```python
tweets[['tweet_text', 'tweet_text_googletrans']]
```

Cool, let's save it back to the CSV.

```python
tweets.to_csv('results/AlinejadMasih.csv')
```
